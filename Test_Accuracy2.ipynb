```python
import docx2txt
import PyPDF2
from pdf2docx import Converter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Store the job description into a variable
job_description = docx2txt.process("open-excelprivatesolution.docx")

# Print the job description
print(job_description)
```

```python
# code to convert pdf into doc
from pdf2docx import Converter

pdf_file = 'Rohit Bijlani.pdf'
docx_file = 'sample.docx'

# convert pdf to docx
cv = Converter(pdf_file)
cv.convert(docx_file, start=0, end=None)
cv.close()
resume = docx2txt.process("sample.docx")
print(resume)
resume = docx2txt.process("open-excelprivatesolution.docx")
```

```python
# A list of text
text = [job_description,resume]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#Print the similarity scores
print("\nSimilarity Scores:")
print(cosine_similarity(count_matrix))
```

```python
#get the match percentage
matchPercentage = cosine_similarity(count_matrix)[0][1] * 100
matchPercentage = round(matchPercentage, 2) # round to two decimal
print("Your resume matches about "+ str(matchPercentage)+ "% of the job description.")
```
